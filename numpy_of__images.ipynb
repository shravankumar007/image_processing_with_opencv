# Image Mask

```
img = img/255.0
nrows, ncols, channels = img.shape
row, col = np.ogrid[:nrows, :ncols]
cnt_row, cnt_col = nrows/2, ncols/2 #finding centroid of the image

outer_disk_mask = ((row - cnt_row) ** 2 + (col - cnt_col) ** 2 > (nrows/2) ** 2) #formula for outer disk
img.setflags(write=1)
img[outer_disk_mask]=0 #assigning outer mask pixel 0

plt.imshow(img)
plt.axis('off')
plt.title('Masked Image')
plt.show()   
```

# image channels

```
r = img[:, :, 0]
g = img[:, :, 1]
b = img[:, :, 2]

output = [img, r, g, b]

titles = ['Image', 'Red', 'Green', 'Blue']

for i in range(4):
    plt.subplot(2, 2, i+1)
    
    plt.axis('off')
    plt.title(titles[i])
    if i == 0:
        plt.imshow(output[i])
    else:
        plt.imshow(output[i], cmap='gray')

plt.show()
```

in the above picture we can see red channel image has high white pixel means the orginal image has highest red pixel

# Logical operatrions

```
img = plt.imread('/content/2216849948.jpg')
img2 = plt.imread('/content/lena.jpg')
img = cv2.resize(img,(350,350))
plt.imshow(np.bitwise_and(img, img2))
plt.show()
```

```
plt.imshow(np.bitwise_or(img2, img))
plt.show()
```

```
plt.subplot(1, 2, 1)
plt.imshow(np.bitwise_xor(img, img2))
plt.subplot(1, 2, 2)
plt.imshow(np.bitwise_xor(img2, img))
plt.show()
```

```
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.subplot(1, 2, 2)
plt.imshow(np.bitwise_not(img))
plt.show()
```


# Image Histograms with NumPy and Matplotlib

```
r = img[:, :, 0]
g = img[:, :, 1]
b = img[:, :, 2]
```

```
plt.subplots_adjust(hspace=0.5, wspace=0.5)

plt.subplot(2, 2, 1)
plt.title('Original Image')
plt.imshow(img)

hist, bins = np.histogram(r.ravel(), bins=256, range=(0, 256))
plt.subplot(2, 2, 2)
plt.title('Red Histogram')
plt.bar(bins[:-1], hist)

hist, bins = np.histogram(g.ravel(), bins=256, range=(0, 256))
plt.subplot(2, 2, 3)
plt.title('Green Histogram')
plt.bar(bins[:-1], hist)

hist, bins = np.histogram(b.ravel(), bins=256, range=(0, 256))
plt.subplot(2, 2, 4)
plt.title('Blue Histogram')
plt.bar(bins[:-1], hist)

plt.show()
```
