# image reading with **matplotlib**

```
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```
img = plt.imread('lena.jpg')
plt.imshow(img)
plt.axis('off')
plt.title('lena')
```

```
plt.imsave('lenamatplot.jpg',img)
```

```
print(type(img))
print(img.shape)
print(img.ndim)
print(img.size)

print(img.dtype)
print(img.nbytes)
```

# image reading with **scikitlearn image**

```
from skimage import io
import skimage
import os
filename = os.path.join(skimage.data_dir, 'camera.png')
camera = io.imread(filename)
plt.figure(figsize=(4, 4))
plt.imshow(camera, cmap='gray', interpolation='nearest')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```
io.imsave('skimage_camera.jpg', camera)
```
