```python
import cv2
import matplotlib.pyplot as plt
```

```python
imgpath = "/content/lena.jpg"
img = cv2.imread(imgpath, 0)
```

```python
plt.imshow(img)
plt.title('Greyscale Image - Default Colormap')
plt.show()
```

```python
plt.imshow(img, cmap = 'gray')
plt.title('Greyscale Image - Grey Colormap')
plt.show()
```

```python
imgpath = "/content/lena.jpg"
img = cv2.imread(imgpath, 1)
```

```python
plt.imshow(img)
plt.title('Color Image BGR')
plt.show()
```

```python
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
plt.title('Color Image BGR')
plt.show()
```

```python
j = 0
for filename in dir(cv2):
    if filename.startswith('COLOR_'):
        print(filename)
        j =  j +1
```
